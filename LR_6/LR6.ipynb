```python
import altair as alt
from vega_datasets import data
```

```python
d = data.gapminder()
data = d[d['country'].isin(['Turkey', 'Spain', 'Mexico'])]
```

```python
data.head()
```

```python
alt.Chart(data).mark_line().encode(
    x='year:O', 
    y='life_expect:Q', 
    color='country:N'
).properties(
    title='Изменение life_expect по странам'
)
```

```python
alt.Chart(data).mark_bar().encode(
    x=alt.X('life_expect:Q', bin=True, title='Продолжительность жизни'),
    y='count():Q', 
    color='country:N',
    tooltip=['country:N', 'lifeExp:Q', 'count():Q']
).properties(
    title='Распределение life_expect по странам'
)
```

```python
 alt.Chart(data).mark_point().encode(
    x='fertility:Q',
    y='life_expect:Q',
    color='country:N',
    size='pop:Q',
    tooltip=['country:N', 'fertility:Q', 'life_expect:Q', 'pop:Q']
).properties(
    title='Связь fertility и life_expect'
)
```

```python
alt.Chart(data).mark_area().encode(
    x='year:O',
    y='pop:Q',
    color='country:N',
    tooltip=['country:N', 'year:O', 'pop:Q']
).properties(
    title='Динамика изменения pop по годам'
)
```

```python
alt.Chart(data).mark_boxplot().encode(
    x='country:N',  
    y='fertility:Q',  
    color='country:N',
    tooltip=['country:N', 'fertility:Q']
).properties(
    title='Распределение fertility по странам'
)
```

```python
alt.Chart(data).mark_bar().encode(
    x='country:N',
    y='mean(fertility):Q',
    color='country:N',
    tooltip=['country:N', 'mean(fertility):Q']
).properties(
    title='Среднее fertility по странам'
)
```

```python
data_1960 = data[data['year'] == 1960]
alt.Chart(data_1960).mark_arc(innerRadius=50).encode(
    theta='pop:Q', 
    color=alt.Color('country:N', legend=None),
    tooltip=['country:N', 'pop:Q', 'country:N']
).properties(
    title='Население по странам в 1960 году'
)
```

```python
 alt.Chart(data).transform_density(
    'fertility', as_=['fertility', 'density'], groupby=['country']
).mark_line().encode(
    x='fertility:Q',
    y='density:Q',
    color='country:N',
    tooltip=['country:N', 'fertility:Q', 'density:Q']
).properties(
    title='Плотность распределения fertility по странам'
)
```
