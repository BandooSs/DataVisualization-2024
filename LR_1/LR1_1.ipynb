```python
import matplotlib.pyplot as plt
import pandas as pd
from vega_datasets import data
```

```python
d = data.gapminder()
data = d[d['country'].isin(['Turkey', 'Spain', 'Mexico'])]
```

```python
data.head()
```

```python
countries = data['country'].unique()
for country in countries:
    country_data = data[data['country'] == country]
    plt.plot(country_data['year'], country_data['life_expect'], label=country)
plt.title('Изменение life_expect по странам', fontsize=14)
plt.xlabel('Год', fontsize=12)
plt.ylabel('Продолжительность жизни', fontsize=12)
plt.xticks(rotation=45)
plt.legend(title="Страна", bbox_to_anchor=(1.05, 1), loc='upper left')
```

```python
bins = range(int(data['life_expect'].min()), int(data['life_expect'].max()) + 5, 5)

for country in data['country'].unique():
    country_data = data[data['country'] == country]
    plt.hist(country_data['life_expect'], bins=bins, alpha=0.5, label=country)

plt.title('Распределение life_expect по странам', fontsize=14)
plt.xlabel('Продолжительность жизни', fontsize=12)
plt.ylabel('Частота', fontsize=12)
plt.legend(title="Страна", bbox_to_anchor=(1.05, 1), loc='upper left')
```

```python
data['fertility'] = pd.to_numeric(data['fertility'], errors='coerce')
data['life_expect'] = pd.to_numeric(data['life_expect'], errors='coerce')
data['pop'] = pd.to_numeric(data['pop'], errors='coerce')

countries = data['country'].unique()
for country in countries:
    country_data = data[data['country'] == country]
    plt.scatter(
        country_data['fertility'],
        country_data['life_expect'],
        s=country_data['pop'] / 1e6,
        label=country,
        alpha=0.6,
        edgecolors='w',
    )
plt.title('Связь fertility и life_expect')
plt.xlabel('fertility', fontsize=12)
plt.ylabel('life_expect', fontsize=12)
plt.legend(title="country", bbox_to_anchor=(1.05, 1), loc='upper left')  # Легенда
plt.grid(True)
```

```python
data['year'] = pd.to_numeric(data['year'], errors='coerce')
data['pop'] = pd.to_numeric(data['pop'], errors='coerce')

countries = data['country'].unique()
for country in countries:
    country_data = data[data['country'] == country]
    plt.fill_between(country_data['year'], country_data['pop'], label=country, alpha=0.6)
plt.xlabel('year', fontsize=12)
plt.ylabel('pop', fontsize=12)
plt.title('Динамика изменения pop по годам')
plt.xticks(rotation=45)
```

```python
import seaborn as sns
```

```python
data['fertility'] = pd.to_numeric(data['fertility'], errors='coerce')
sns.boxplot(x='country', y='fertility', data=data, palette="Set2")
plt.xlabel('country', fontsize=12)
plt.title('Распределение fertility по странам')
plt.ylabel('fertility', fontsize=12)
plt.xticks(rotation=45)
```

```python
mean_fertility = data.groupby('country')['fertility'].mean().reset_index()
plt.bar(mean_fertility['country'], mean_fertility['fertility'])
plt.title('Среднее fertility по странам')
plt.xlabel('country', fontsize=12)
plt.ylabel('fertility', fontsize=12)
plt.xticks(rotation=45)
```

```python
from vega_datasets import data
d = data.gapminder()
data = d[d['country'].isin(['Turkey', 'Spain', 'Mexico'])]
data_1960 = data[data['year'] == 1960]
population_by_country = data_1960.groupby('country')['pop'].sum()

fig, ax = plt.subplots(figsize=(8, 8))

wedges, texts, autotexts = ax.pie(
    population_by_country, 
    labels=population_by_country.index, 
    autopct='%1.1f%%', 
    startangle=90,
    colors=plt.cm.Paired.colors,
    wedgeprops={'width': 0.3} 
)

for autotext in autotexts:
    autotext.set(fontsize=10, color='black')

plt.title('Население по странам в 1960 году', fontsize=14)
plt.show()
```

```python
countries = d['country'].unique()
for country in countries:
    country_data = data[data['country'] == country]
    sns.kdeplot(
        data=country_data,
        x='fertility',
        label=country,
        linewidth=2  
    )
plt.title('Плотность распределения fertility по странам')
plt.xlabel('fertility', fontsize=12)
plt.ylabel('density', fontsize=12)
plt.legend(title='country', bbox_to_anchor=(1.05, 1), loc='upper left')
```
