```python
import matplotlib.pyplot as plt
import numpy as np
import random
```

## Пример 1
Обычный plot - строит график функции y(x)

```python
n = 20
X = np.arange(n)
y = np.random.rand(n)
fig1, ax1 = plt.subplots()
ax1.plot(X,y,linestyle = '-.', color = 'red', marker = 'd')
ax1.set_title('Первый пример')
ax1.set_xlabel('Ось X')
ax1.set_ylabel('Ось y')
```

## Пример 2
scatter - для нагдяности распределения признаков класса

```python
def train_data_1(k: float = 1.55, b: float = 2.3, args_range: float = 10, rand_range: float = 0.1,
                 n_points: int = 300):
    X = np.zeros((n_points, 2), dtype=float)
    X[:, 0] = np.array([random.uniform(0.0 * args_range, 1.0 * args_range) for _ in range(n_points)])
    X[:, 1] = np.array([random.uniform(0.0 * args_range, 1.0 * args_range) for _ in range(n_points)])
    g = np.array([0 if k * X[i, 0] + b >= 1.5 * X[i, 1] else 1 for i in range(n_points)], dtype = int)
    return X, g
```

```python
X, y = train_data_1()
maskClass0 = y == 0
X0 = X[maskClass0]
X1 = X[~maskClass0]
fig2, ax2 = plt.subplots()
ax2.scatter(X0[:,0],X0[:,1], color = 'red', marker = 'd', label = '1')
ax2.scatter(X1[:,0],X1[:,1], color = 'blue', marker = '*', label = '2')
ax2.legend()
ax2.set_title('Второй пример')
ax2.set_xlabel('Ось X1')
ax2.set_ylabel('Ось X2')
```

## Пример 3
bar - для вывода гистограммы

```python
X, y = train_data_1()

class_counts = np.bincount(y)

fig3, ax3 = plt.subplots()

ax3.bar(range(len(class_counts)), class_counts, color=['red', 'blue'], alpha=0.7)

ax3.set_xticks(range(len(class_counts)))  # Установка меток на оси X
ax3.set_xticklabels(['Класс 0', 'Класс 1'])  # Подписи для классов
ax3.set_title('Второй пример')
ax3.set_xlabel('Класс')
ax3.set_ylabel('Кол-во элементов класса')
```

## Пример 4
contour - для взгяда сверху на функцию

```python
func  = lambda x, y: x**2 + y**2
x = np.linspace(-20,20,500)
y = np.linspace(-20,20,500)
X, Y = np.meshgrid(x, y)
Z = func (X, Y)
fig4, ax4 = plt.subplots()
contour = ax4.contour(X, Y, Z,  levels=800, cmap='viridis')
fig4.colorbar(contour, ax=ax4)
ax4.set_title('Четвертый пример')
ax4.set_xlabel('Ось x')
ax4.set_ylabel('Ось y')
```

## Пример 5
plot_surface - 3d график

```python
from mpl_toolkits.mplot3d import Axes3D
func  = lambda x, y: x**2 + y**2
x = np.linspace(-20,20,500)
y = np.linspace(-20,20,500)
X, Y = np.meshgrid(x, y)
Z = func (X, Y)
fig5, ax5 = plt.subplots()
ax5 = fig5.add_subplot(111, projection='3d')
surface = ax5.plot_surface(X, Y, Z, cmap='viridis')
fig5.colorbar(surface, ax=ax5, shrink=0.5, aspect=5)
ax5.set_title('Пятый пример')
ax5.set_xlabel('Ось x')
ax5.set_ylabel('Ось y')
```

## Пример 6
pie -  круговая диаграмма

```python
X, y = train_data_1()

# Подсчет количества элементов в каждом классе
class_counts = np.bincount(y)
labels = [f'Класс 0: {class_counts[0]} ({class_counts[0] / sum(class_counts) * 100:.1f}%)',
          f'Класс 1: {class_counts[1]} ({class_counts[1] / sum(class_counts) * 100:.1f}%)']
fig6, ax6 = plt.subplots()
ax6.pie(class_counts, labels=labels, autopct='%1.1f%%', startangle=90, colors=['red', 'blue'])
ax6.set_title('Шестой пример')
```
